Python wraper to compute TM Score

```python
import subprocess
import re

def compute_tm_score(pdb1, pdb2, tmscore_path="./TMscore"):
    result = subprocess.run(
        [tmscore_path, pdb1, pdb2],
        capture_output=True,
        text=True
    )
    stdout_text = result.stdout
    #print("result:", stdout_text)
    match = re.search(
        r"TM-score\s*=\s*([0-9]*\.?[0-9]+)",
        stdout_text
    )
    #print(match)
    if not match:
        raise ValueError("TM-score not found in output")
    return float(match.group(1))

```
